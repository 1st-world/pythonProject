```python
# 배열을 순회하면서 간단하게 처리할 수 있게 하는 map 함수
my_list = [1, 2, 3, 4, 5]
# -> [2, 3, 4, 5, 6]

def add_one(n):
    return n + 1

result = list(map(add_one, my_list))
print(result)
```

```python
def rank(score):
    result = []
    # 1. 평균값을 가지는 리스트
    avg = []
    for i, j in score:
        avg.append((i + j) / 2)
    # print(avg)
    # -> [75.0, 70.0, 55.0, 65.0]

    # 2. 평균값을 내림차순으로 정렬
    avg_sort = sorted(avg, reverse=True)
    # print(avg_sort)
    # -> [75.0, 70.0, 65.0, 55.0]

    # 3. 비교해서 등수 처리
    for i in avg:
        result.append(avg_sort.index(i) + 1)
    # print(result)
    # [1, 2, 4, 3]
    return result
```

```python
# 소수 판별 함수

def is_prime(num_list):
    result = []
    for num in num_list:
        isPrime = True
        for i in range(2, num):
            if num % i == 0:
                isPrime = False
                break
        result.append(isPrime)
    return result

is_prime([2, 3, 4, 5, 6, 7, 8, 9])
```

### Class

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def walk(self):
        print(f"{self.name}이(가) 걸어가요.")
    def sleep(self):
        print(f"{self.name}이(가) 잠을 자요.")
```

```python
person1 = Person("박지민", 31)
person1.walk()
person1.sleep()
```

```python
class NewPerson(Person):
    def __init__(self, name, age):
        super().__init__(name, age)
    def fly(self):
        print(f"{self.name}이(가) 하늘을 날아요!")
    def walk(self):     # Override
        print(f"{self.name}이(가) 너무 잘 걸어가요!")
```

```python
person2 = NewPerson("정호석", 33)
person2.fly()
person2.walk()
person2.sleep()
```
